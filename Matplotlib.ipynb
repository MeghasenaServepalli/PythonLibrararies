```python
#import all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(10,30)
y=np.arange(20,40)
```

```python
a=np.arange(40,55)
b=np.arange(45,60)
```

```python
#ploting using matplotlib
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("2D Graph")
plt.show()
plt.savefig("Test1.png")
```

```python
plt.plot(x,y,c='g')
```

```python
# the dimensions of a and b have to same means the shape (no of elements have to be same)
plt.plot(a,b,'ro',linestyle='-.',linewidth=2,markersize=13)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Diagram')
```

## creating subplots

```python
plt.title("plots",pad=20)
plt.subplot(221)
#plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.plot(x,y,'r:',c='b',linewidth=3,markersize=14)
plt.subplot(2,2,2)
plt.plot(x,y,'-.',c='g')
plt.subplot(2,2,3)
plt.scatter(x,y)
#plt.subplots_adjust(hspace=5, wspace=5)
#plt.tight_layout(h_pad=0.5, w_pad=0.5)
plt.subplot(2,2,4)
plt.plot(x,y)

```

```python
#perform sine curve
x=np.arange(0,6*np.pi,0.1)
y=np.sin(x + 1)
plt.figure(figsize=(10,5))
plt.title("sine wave curve")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x,y)
```

```python
#get x,y points for sine and cosine curves
x=np.arange(0,4*np.pi,0.1)
y1_sin=np.sin(x)
y2_cosine=np.cos(x)
#setting up the subplots for sine
#plt.title("sine")
plt.subplot(2,1,1)
plt.plot(x,y1_sin,'r--')
plt.title("sine")
#plot for cosine
plt.subplot(212)
plt.plot(x,y2_cosine,'c:')
plt.title("cosine")
plt.subplots_adjust()
plt.tight_layout(h_pad=0.5, w_pad=0.5)
```

```python
#Bar plots
x=[2,5,7]
y=[9,11,15]
x1=[1,5,10]
x2=[4,7,9]
plt.bar(x,y,color='r',align='edge',width=1,edgecolor='k', tick_label=None)
plt.bar(x1,x2,color='c')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('bar plot')
plt.show()
```

```python
# Histograms
a=np.array([20,9,7,12,16,19,22,27,39])
plt.hist(a,bins=8)
```

```python
#Box Plot with Matplotlib
dat=[np.random.normal(0,std,100) for std in range(1,6)]
plt.boxplot(dat,vert=True,patch_artist=True,labels=['A','B','C','D','E'])
plt.show()
```

```python
#Pie chart
labels='CS','IT','MIS','CIVIL','MECH'
sizes=[100,150,200,170,150]
colors=['cyan','magenta','yellow','blue','green']
explode=(0.2,0,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.2f%%',shadow=True,startangle=90, )
plt.axis('equal')
plt.show()
```
