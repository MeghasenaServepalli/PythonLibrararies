## Seaborn Tutorial
## Distribution Plots

# Dataset Tips dataset

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.head()
```

```python
df.dtypes
```

# Correlation with Heatmap
# A Corr heatmap uses colored cells, typically in monochrome scale to show 2D corr matrix(table) between two discrete dim or event types. used more in feature selection

```python

new_df=df.select_dtypes(include='number')
new_df.corr()
```

```python

sns.heatmap(new_df.corr())
```

## Joint Point
# It allows us to study the relationship between 2 num varibales. so the middle chart shows correlation. Its a scatterplot in general, have a hexbin plot, 2D diagram or 2D plot

```python
sns.jointplot(x='total_bill',y='size',data=df,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='size',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='size',data=df,hue='tip')
```

```python
sns.jointplot(x='total_bill',y='size',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='size',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='size',data=df,hue='tip',palette='deep')
```

## Pair Plot
# also known as scatter plot, in which 1 variable in the same data row is matched with another variable's value, shows all variables paired with all the other variables.



```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker',kind='kde')
```

```python
sns.pairplot(df,hue='sex',kind='reg')
```

```python
sns.pairplot(df,hue='smoker',kind='kde',palette='deep')
```

```python
sns.pairplot(df,hue='size',kind='hist',palette='deep')
```

```python
sns.pairplot(df,hue='size',kind='scatter',palette='deep')
#here kind can't be resid and hex
```

```python
df['size'].value_counts()
```

```python
df['smoker'].value_counts()
```

## Dist Plot
# checks the distribution of column features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['total_bill'],kde=False)
```

```python
sns.distplot(df['total_bill'],kde=False,bins=5,vertical=True)
```

```python
sns.histplot(df['total_bill'],kde=False,bins=5,palette='deep')
```

```python
sns.histplot(df,x=df['total_bill'],y=df['tip'],kde=True,hue='sex',palette='deep')
```

```python
sns.rugplot(df,x=df['total_bill'],y=df['tip'],hue='sex',palette='deep')
```

```python
sns.ecdfplot(df,x=df['total_bill'],hue='sex',palette='deep')
#emperical cummulative distribution function cant use 2 axis like x,y
```

## Categorical plots
# used on categorical variables like sex, true or false values, strings
# types: 1 BOX PLOT 2. VIOLIN PLOT 3. COUNTPLOT 4. BAR PLOT

```python
## Count Plot
sns.countplot(df,x=df['sex'])
```

```python
sns.countplot(y=df['smoker'])
```

```python
sns.countplot(df)
```

```python
sns.countplot(df,x=df['sex'],hue='smoker')
```

## Bar Plot

```python
sns.barplot(df,x=df['total_bill'],y=df['sex'],hue='size')
```

```python
df.head()
```

##Box Plot
# Also called Whisker plot is a graph that presents information from a five-number summary.

```python
sns.boxplot(df,x=df['smoker'],y=df['total_bill'],hue='sex')
```

```python
sns.boxplot(df,x=df['smoker'],y=df['total_bill'],hue='sex',palette='rainbow')
```

```python
sns.boxplot(df,x=df['smoker'],y=df['total_bill'],hue='sex',palette='rainbow',orient='y')
```

## Violin Plot: helps us see both distribution of data in terms of KDE and Box plot

```python
sns.violinplot(df,x=df['total_bill'],y=df['day'])
```

```python
sns.violinplot(df,x=df['total_bill'],y=df['day'],hue='smoker')
```

```python
sns.violinplot(df,x=df['total_bill'],y=df['day'],hue='smoker',palette='rainbow')
```
