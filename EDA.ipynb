```python
##now lets perform data cleaning
#we can see weather passengers in higherclasses tend to be old , we will use avergae age values to impute based
#on pclass for Ages
def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 38
    elif Pclass==2:
      return 29
    else:
      return 25
  else:
    return Age

```

```python
## apply this function
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
train.drop('Cabin',inplace=True,axis=1)
```

```python
sns.heatmap(train.isnull(),cbar=False,cmap='viridis',yticklabels=False)
```

```python
sns.boxplot(x='Embarked',y='Age',data=train)
```

```python
def impute_Embark(cols):
  Age=cols[1]
  Embarked=cols[0]
  if pd.isnull(Embarked):
    if Age>=22 and Age<=26 :
      return 'S'
    elif Age>=27 and Age<=30:
      return 'C'
    else:
      return 'Q'
  else:
    return Embarked
```

```python
## apply this function
train['Embarked']=train[['Embarked','Age']].apply(impute_Embark,axis=1)
```

```python
## Converting Categorical Features
#Perform 1 hot encoding, get_dummies converts categorical to dummy or indicator variables.
Embarked=pd.get_dummies(train['Embarked'],drop_first=True,dtype=int)
```

```python
pd.get_dummies(train['Embarked'],drop_first=True,dtype=int).head()
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True,dtype=int)
```

```python
sex
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train
```

```python
train=pd.concat([train,sex,Embarked],axis=1)
```

```python
train
```

```python
## Perform Logistic Regression
train.drop('Survived',axis=1).head()
```

```python
train['Survived'].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived',axis=1),
                                                    train['Survived'], test_size=0.30,
                                                    random_state=101)
```

```python
## Training and Predicting
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
y_pred=logmodel.predict(X_test)
```

```python
y_pred
```

```python
##get the confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```
