# reading different data sources with the help of pandas

```python
from io import StringIO,BytesIO
data=('col1,col2,col3\n'
'x,y,1\n'
'a,b,2\n'
'c,d,3')
```

```python
type(data)
```

```python
StringIO()
```

```python
df3=pd.read_csv(StringIO(data))
df3
```

```python
#usage of columns
df4=pd.read_csv(StringIO(data),usecols=['col1','col3'])
```

```python
df4
```

```python
df5=pd.read_csv(StringIO(data),usecols=lambda x:x.upper() in ['COL1','COL3'])
```

```python
df5
```

```python
#takes object as string
df6=pd.read_csv(StringIO(data),dtype=object)
```

```python
df6
```

```python
df6=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'int64'})
```

```python
df6
```

```python
#dont take values in index-index_col=false
df7=pd.read_csv(StringIO(data),index_col=False)
```

```python
df7
```

```python
df7['col1'][0]
```

```python
df7.dtypes
```

```python
#quoting and escape characters --more used in NLP
data='a,b\n"hello,\\"Bob\\",nice to see you",5'
pd.read_csv(StringIO(data),escapechar='\\')
```

```python
#URL to CSV
import pandas as pd
import requests

url = 'https://download.bls.gov/pub/time.series/cu/cu.item'
response = requests.get(url)
data = StringIO(response.text)
df8 = pd.read_csv(data, sep='\t')
```
